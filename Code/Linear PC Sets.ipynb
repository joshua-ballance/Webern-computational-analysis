```python
## set hyperparameters
max_duration = 5 #this is the maximum permitted duration from the start of the first note to the start of the last
max_rest_length = 2 #this is the maximum permitted duration for an intra-unit rest
min_harmony_size = 3 #this is the minimum permitted number of notes to constitute a harmonic unit
corpus_complete_harmonies = []
for a in corpus_complete_list_of_parts:
    fileTitle = a[0]
    piece_harmonies = [fileTitle]
    piece_info = a[1:]
    for b in range(0,len(piece_info)): #i.e. for each part
        part_info = piece_info[b]
        part_number = part_info[0][0]
        link_info = part_info[0][1:]
        notes_and_rests_in_current_part = part_info[1]
        this_part_harmonies = [b, link_info] 
        do_analysis()
        for el in link_info:
            if el[0] == 'no_link':
                pass
            else:
                for item in el:
                    if item[0] == 'start_link':
                        #this does the analysis for standard entries
                        part_to_link_to_number = item[1] 
                        #this isolates and then organises the other part
                        part_to_link_to = piece_info[part_to_link_to_number]
                        part_to_link_to_notes_and_rests = part_to_link_to[1]
                        current_part_first_note = []
                        for m in notes_and_rests_in_current_part:
                            if 'Rest' in str(m[0]):
                                pass
                            else:
                                current_part_first_note.append(m)
                        current_part_first_note = current_part_first_note[0]
                        current_part_first_note_offset = current_part_first_note[0].offset
                        current_part_first_note_position = notes_and_rests_in_current_part.index(current_part_first_note)
                        #this isolates the data for the first note of the part to be analysed, which is the linking note 
                        for n in range(0, len(part_to_link_to_notes_and_rests)):
                            note_to_try = part_to_link_to_notes_and_rests[n]
                            if current_part_first_note_offset == note_to_try[0].offset:
                                part_to_link_to_start_note = part_to_link_to_notes_and_rests[n]
                                #this creates a list of positions in the part to link to which are in reverse order,
                                #and then checks if adding them on to the list will make harmonies that are the right length
                                for o in reversed(range(0, n-1)):
                                    if 'Rest' in str(part_to_link_to_notes_and_rests[o]):
                                        pass
                                    else:
                                        collection_of_notes = []
                                        for p in range(o, n):
                                            collection_of_notes.append(part_to_link_to_notes_and_rests[p])
                                        collection_of_notes_seconds = []
                                        for q in collection_of_notes:
                                            collection_of_notes_seconds.append(q[1])
                                        if sum(collection_of_notes_seconds) < max_duration:
                                            for r in range(current_part_first_note_position, len(notes_and_rests_in_current_part)):###
                                                if sum(collection_of_notes_seconds) < max_duration:
                                                    collection_of_notes.append(notes_and_rests_in_current_part[r])
                                                    collection_of_notes_seconds.append(notes_and_rests_in_current_part[r][1])
                                                    list_of_items_to_consider = collection_of_notes
                                                    if 'Rest' in str(list_of_items_to_consider[len(list_of_items_to_consider)-1]):
                                                        pass
                                                    else:
                                                        lists_of_consecutive_rest_durations = []
                                                        #this isolates rests in the harmonic segment and checks whether their
                                                        #cumulative duration is the right length
                                                        for g in range(1, len(list_of_items_to_consider)):
                                                            temp_first_item = list_of_items_to_consider[g]
                                                            temp_list_of_rests = []
                                                            if 'Rest' in str(temp_first_item):
                                                                temp_list_of_rests.append(temp_first_item)
                                                                for h in range(g+1,len(list_of_items_to_consider)):
                                                                    if 'Rest' in str(list_of_items_to_consider[h]):
                                                                        temp_list_of_rests.append(list_of_items_to_consider[h])
                                                                    else:
                                                                        break
                                                            temp_list_of_rest_durations = []
                                                            for i in temp_list_of_rests:
                                                                temp_list_of_rest_durations.append(i[1])
                                                            lists_of_consecutive_rest_durations.append(sum(temp_list_of_rest_durations))
                                                        whether_to_proceed = []
                                                        for j in lists_of_consecutive_rest_durations:
                                                            if j > max_rest_length:
                                                                whether_to_proceed.append('no')
                                                        if len(whether_to_proceed) > 0:
                                                            pass
                                                        #this removes any duplicate items
                                                        else:
                                                            tidy_list_of_items_to_consider = []
                                                            items_to_delete = []
                                                            for s in range(0, len(list_of_items_to_consider)):
                                                                for t in range(0, len(list_of_items_to_consider)):
                                                                    if list_of_items_to_consider[s][0].offset == list_of_items_to_consider[t][0].offset:
                                                                        if s == t:
                                                                            pass
                                                                        else:
                                                                            if 'Rest' in str(list_of_items_to_consider[s]):
                                                                                items_to_delete.append(s)
                                                                            else:
                                                                                if 'Rest' in str(list_of_items_to_consider[t]):
                                                                                    items_to_delete.append(t)
                                                                                else:
                                                                                    if list_of_items_to_consider[s][0].pitch.pitchClass == list_of_items_to_consider[t][0].pitch.pitchClass:
                                                                                        items_to_delete.append(s)
                                                            for u in range(0, len(list_of_items_to_consider)):
                                                                if u in items_to_delete:
                                                                    pass
                                                                else:
                                                                    tidy_list_of_items_to_consider.append(list_of_items_to_consider[u])
                                                            #this converts pitch material into pitch classes
                                                            list_of_notes_to_analyse = []
                                                            for k in tidy_list_of_items_to_consider:
                                                                if 'Rest' in str(k):
                                                                    pass
                                                                else:
                                                                    list_of_notes_to_analyse.append(k[0])
                                                            #this checks the harmonic unit is big enough
                                                            if len(list_of_notes_to_analyse) < min_harmony_size:
                                                                pass
                                                            else:
                                                                chord_to_analyse = chord.Chord(list_of_notes_to_analyse)
                                                                this_part_harmonies.append(chord_to_analyse.primeForm)
                    if item[0] == 'end_link':
                        part_to_link_to_number = item[1]
                        #this isolates and then organises the other part
                        part_to_link_to = piece_info[part_to_link_to_number]
                        part_to_link_to_notes_and_rests = part_to_link_to[1]
                        current_part_last_note = []
                        for m in reversed(notes_and_rests_in_current_part):
                            if 'Rest' in str(m[0]):
                                pass
                            else:
                                current_part_last_note.append(m)
                        current_part_last_note = current_part_last_note[0]
                        current_part_last_note_offset = current_part_last_note[0].offset
                        current_part_last_note_position = notes_and_rests_in_current_part.index(current_part_last_note)
                        #this isolates the data for the first note of the part to be analysed, which is the linking note 
                        for n in range(0, len(part_to_link_to_notes_and_rests)):
                            note_to_try = part_to_link_to_notes_and_rests[n]
                            if current_part_last_note_offset == note_to_try[0].offset:
                                part_to_link_to_start_note = part_to_link_to_notes_and_rests[n]
                                #this creates a list of positions in the part to link to which are in reverse order,
                                #and then checks if adding them on to the list will make harmonies that are the right length
                                #this creates a list of positions in the part to link to which are in reverse order,
                                #and then checks if adding them on to the list will make harmonies that are the right length
                                for o in range(n+1, len(part_to_link_to_notes_and_rests)):
                                    if 'Rest' in str(part_to_link_to_notes_and_rests[o]):
                                        pass
                                    else:
                                        collection_of_notes = []
                                        for p in range(n, o):
                                            collection_of_notes.append(part_to_link_to_notes_and_rests[p])
                                        collection_of_notes_seconds = []
                                        for q in collection_of_notes:
                                            collection_of_notes_seconds.append(q[1])
                                        if sum(collection_of_notes_seconds) < max_duration:
                                            for r in range(0, current_part_last_note_position):###
                                                if sum(collection_of_notes_seconds) < max_duration:
                                                    collection_of_notes.append(notes_and_rests_in_current_part[r])
                                                    collection_of_notes_seconds.append(notes_and_rests_in_current_part[r][1])
                                                    list_of_items_to_consider = collection_of_notes
                                                    if 'Rest' in str(list_of_items_to_consider[len(list_of_items_to_consider)-1]):
                                                        pass
                                                    else:
                                                        lists_of_consecutive_rest_durations = []
                                                        #this isolates rests in the harmonic segment and checks whether their
                                                        #cumulative duration is the right length
                                                        for g in range(1, len(list_of_items_to_consider)):
                                                            temp_first_item = list_of_items_to_consider[g]
                                                            temp_list_of_rests = []
                                                            if 'Rest' in str(temp_first_item):
                                                                temp_list_of_rests.append(temp_first_item)
                                                                for h in range(g+1,len(list_of_items_to_consider)):
                                                                    if 'Rest' in str(list_of_items_to_consider[h]):
                                                                        temp_list_of_rests.append(list_of_items_to_consider[h])
                                                                    else:
                                                                        break
                                                            temp_list_of_rest_durations = []
                                                            for i in temp_list_of_rests:
                                                                temp_list_of_rest_durations.append(i[1])
                                                            lists_of_consecutive_rest_durations.append(sum(temp_list_of_rest_durations))
                                                        whether_to_proceed = []
                                                        for j in lists_of_consecutive_rest_durations:
                                                            if j > max_rest_length:
                                                                whether_to_proceed.append('no')
                                                        if len(whether_to_proceed) > 0:
                                                            pass
                                                        #this removes any duplicate items
                                                        else:
                                                            tidy_list_of_items_to_consider = []
                                                            items_to_delete = []
                                                            for s in range(0, len(list_of_items_to_consider)):
                                                                for t in range(0, len(list_of_items_to_consider)):
                                                                    if list_of_items_to_consider[s][0].offset == list_of_items_to_consider[t][0].offset:
                                                                        if s == t:
                                                                            pass
                                                                        else:
                                                                            if 'Rest' in str(list_of_items_to_consider[s]):
                                                                                items_to_delete.append(s)
                                                                            else:
                                                                                if 'Rest' in str(list_of_items_to_consider[t]):
                                                                                    items_to_delete.append(t)
                                                                                else:
                                                                                    if list_of_items_to_consider[s][0].pitch.pitchClass == list_of_items_to_consider[t][0].pitch.pitchClass:
                                                                                        items_to_delete.append(s)
                                                            for u in range(0, len(list_of_items_to_consider)):
                                                                if u in items_to_delete:
                                                                    pass
                                                                else:
                                                                    tidy_list_of_items_to_consider.append(list_of_items_to_consider[u])
                                                            #this converts pitch material into pitch classes
                                                            list_of_notes_to_analyse = []
                                                            for k in tidy_list_of_items_to_consider:
                                                                if 'Rest' in str(k):
                                                                    pass
                                                                else:
                                                                    list_of_notes_to_analyse.append(k[0])
                                                            #this checks the harmonic unit is big enough
                                                            if len(list_of_notes_to_analyse) < min_harmony_size:
                                                                pass
                                                            else:
                                                                chord_to_analyse = chord.Chord(list_of_notes_to_analyse)
                                                                this_part_harmonies.append(chord_to_analyse.primeForm)
        piece_harmonies.append(this_part_harmonies) 
    corpus_complete_harmonies.append(piece_harmonies)
```
